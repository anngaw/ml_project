### Zamiana na wartości nominalne
Do grupowania danych konieczna jest zamiana wartości nomianlnych na wartości liczbowe.

Do zamiany (w obie strony) służą poniższe funkcje:

```python
from sklearn import preprocessing
le_data = preprocessing.LabelEncoder()
def to_categorical(column):
    
    column = le_data.fit_transform(column)
    return column
    
def from_categorical(column):
    
    column = le_data.inverse_transform(column)
    return column

data_csv.q0 = to_categorical(data_csv.q0)
data_csv.q1 = to_categorical(data_csv.q1)
data_csv.q2 = to_categorical(data_csv.q2)
data_csv.q3 = to_categorical(data_csv.q3)
data_csv.q4 = to_categorical(data_csv.q4)
data_csv.q5 = to_categorical(data_csv.q5)
data_csv.q6 = to_categorical(data_csv.q6)
```
